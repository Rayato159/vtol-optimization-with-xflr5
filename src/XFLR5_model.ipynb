```python
import numpy as np
import pandas as pd

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import relu
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.optimizers import Adam

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../data/100parashuffle.csv")
df.head()
```

```python
X = df[["Span_w", "Root_Chord_w", "Tip_Chord_w", "Span_e", "Root_Chord_e", "Span_f", "Root_Chord_f", "Tip_Chord_f", "X_f"]]
y = df[["cl", "cd", "cm"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
NN_model = Sequential()

NN_model.add(Dense(units=128, activation=relu, input_shape=(9,)))
NN_model.add(Dense(units=64, activation=relu))
NN_model.add(Dense(units=64, activation=relu))
NN_model.add(Dense(units=32, activation=relu))
NN_model.add(Dense(units=3))

NN_model.compile(optimizer=Adam(), loss=MeanSquaredError())
```

```python
NN_model.fit(X_train, y_train, epochs=600)
```

```python
NN_model.evaluate(X_test, y_test)
```

```python
NN_model.save('XFLR5_model')
```
